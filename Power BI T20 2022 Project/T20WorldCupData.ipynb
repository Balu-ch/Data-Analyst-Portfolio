```python
# importing libraries
import pandas as pd
import json
```

# Match Summary

```python
with open(r'D:\Data Analyst Portfolio\_Projects\Power BI T20 2022 Project\Jsons\t20_wc_match_results.json') as f:
    data =json.load(f)

#storing the data in dataframe    
df_match =pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match.head()
```

```python
#creating a dictionary for mapping the data from Match summary

match_ids_dict ={}

for index, row in df_match.iterrows():
    key1 = row['team1']+' Vs ' +row['team2']
    key2 = row['team2']+' Vs ' +row['team1']
    match_ids_dict[key1]=row['match_id']
    match_ids_dict[key2]=row['match_id']
```

```python
df_match.to_csv(r'D:\Data Analyst Portfolio\_Projects\Power BI T20 2022 Project\Csv\matches.csv' ,index =False)
```

# Batting Summary

```python
#loading the file and storing in the dataframe
with open(r'D:\Data Analyst Portfolio\_Projects\Power BI T20 2022 Project\Jsons\t20_wc_batting_summary.json') as f:
    data =json.load(f)
all_records =[]
    
for rec in data:
    all_records.extend(rec['battingSummary'])
               
df_batting =pd.DataFrame(all_records)
df_batting.head(20)

```

```python
# creating a new column 
df_batting["out/not_out"] =df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting.head(10)
```

```python
#dropping Unnecssary columns
df_batting.drop(columns =['dismissal'],inplace =True)
df_batting.head(10)
```

```python
# cleaning weird symbols from the names
df_batting['batsmanName'] =df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
df_batting['batsmanName'] =df_batting['batsmanName'].apply(lambda x: x.replace('\xa0',''))
df_batting['batsmanName'] =df_batting['batsmanName'].apply(lambda x: x.replace('(c)',''))
df_batting.head(10)
```

```python
#checking if the match has substrings of team1 and team2

for m in range(len(df_match.team1)):
    for b in range(len(df_batting.match)):
        if (df_match.team1[m] in df_batting.match[b] and df_match.team2[m] in df_batting.match[b]):
            #if both team1 and team2 present in the match store them in df_batting
            df_batting['match_id'][b] =df_match['match_id'][m]

df_batting.tail(20)
```

```python
df_batting.to_csv(r'D:\Data Analyst Portfolio\_Projects\Power BI T20 2022 Project\Csv\batting.csv',index =False)
```

# Bowling Summary

```python
with open(r'D:\Data Analyst Portfolio\_Projects\Power BI T20 2022 Project\Jsons\t20_wc_bowling_summary.json') as f:
    data =json.load(f)
all_records =[]
    
for rec in data:
    all_records.extend(rec['bowlingSummary'])
               
df_bowling =pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head(20)
```

```python
#appending match_id Column to the df_bowling
df_bowling['match_id'] =df_bowling['match'].map(match_ids_dict)

df_bowling.head()

```

```python
#saving into the file
df_bowling.to_csv(r'D:\Data Analyst Portfolio\_Projects\Power BI T20 2022 Project\Csv\bowling.csv',index =False)
```
